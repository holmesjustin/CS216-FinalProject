```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
# 2013 teams

data13 = pd.read_csv("Clean Scraped Data/CLEANheight-exp2013.csv")
data13 = data13.drop("Unnamed: 0", axis = 1)
data13.head()
data13["Postseason"] = 0
cbb13 = pd.read_csv("Kaggle Datasets/cbb13.csv")
cbb13 = cbb13[:68]
tournament_teams13 = cbb13["TEAM"].values.tolist()
tournament_teams13 = [tournament_teams13[i].lower() for i in range(len(tournament_teams13))]
for i in range(len(tournament_teams13)):
    tournament_teams13[i] = tournament_teams13[i].replace("st.", "state")
for i in range(len(tournament_teams13)):
    if tournament_teams13[i] == "unlv":
        tournament_teams13[i] = "nevada las vegas"
    elif tournament_teams13[i] == "vcu":
        tournament_teams13[i] = "virginia commonwealth"
    elif tournament_teams13[i] == "saint mary's":
        tournament_teams13[i] = "saint marys ca"
    elif tournament_teams13[i] == "north carolina a&t":
        tournament_teams13[i] = "north carolina at"
    elif tournament_teams13[i] == "liu brooklyn":
        tournament_teams13[i] = "long island university"
    elif tournament_teams13[i] == "albany":
        tournament_teams13[i] = "albany-ny"
tournament_teams13.remove("james madison")
tournament_teams13.remove("liberty")
data13 = data13.set_index("School")
data13.loc[tournament_teams13, "Postseason"] = 1
data13.head(10)
```
