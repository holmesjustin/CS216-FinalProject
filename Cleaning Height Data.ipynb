```python
import pandas as pd
import seaborn as sns
```

```python
#2013

df13 = pd.read_csv("216 project/scraped data/height-exp2013.csv")

df13.columns = ["School", 'Avg_Height','Avg_Experience']
#pd.set_option('display.max_rows', None)


#dropped the following schools due to incomplete data: central connecticut state, connecticut, drexel, 
#james madison, liberty, louisiana lafayette, mississippi valley state, north texas, ohio, southern utah, wagner
df13 = df13.drop([42, 57, 70, 128, 138, 143, 171, 200, 211, 269, 327])
df13
```

```python
df13["Avg_Height"] = df13["Avg_Height"].str.replace("6-", "")
df13["Avg_Height"] = df13["Avg_Height"].astype(float) + 72 
df13
```

```python
df13.to_csv('/Users/Donald/Desktop/CLEANheight-exp2013')
```
